```python
import pandas as pd
from scipy.stats import hmean, lognorm
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("mango-hbox.tsv",sep="\t")
```

```python
df[:5]
```

```python
def pct_at_time(df, t, window=2.1):
    near = df[abs(df['time'] - t) < window]
    
    best_pct = 0
    best_conf = 0
    for pct, pts in near.groupby('mango_pct'):
        if any(pts['mango_conf'] > 0):
            conf = hmean(pts['mango_conf']) * len(pts)/(window * 2 + 1)
        else:
            conf = 0
            
        if conf > best_conf:
            best_conf = conf
            best_pct = pct
    
    return (t, best_pct, best_conf * -100)
```

```python
def rec_pct_at_time(df, window=2.1):
    min(df['time']),max(df['time']))
```

```python
rec_pct_at_time(df)
```

```python
data = [pct_at_time(df, t, window=2.1) for t in range(0, 700)]
data = pd.DataFrame(data, columns=("time", "pct", "conf"))
plt.plot('time', 'pct', data=data)
plt.plot('time', 'conf', data=data)
```
